```python
def loc(df):
    df = df.assign(color='darkgreen')
    df.loc[df['Alert Level'] == 'Level 1', 'color'] = 'seagreen'
    df.loc[df['Alert Level'] == 'Level 2', 'color'] = 'gold'
    df.loc[df['Alert Level'] == 'Level 3', 'color'] = 'red'
    df.loc[df['Alert Level'] == 'Level 3', 'color'] = 'firebrick'
    return df
```
